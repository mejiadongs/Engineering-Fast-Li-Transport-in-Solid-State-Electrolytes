```python
import os
from ase.db import connect
from ase.build import add_adsorbate
from ase.io import read

db_original = connect('initial_db.db')
db_ads = connect('initial_db_Li.db')

adsorbate = 'Li'
z_ads_1 = 2.5

for row in db_original.select():
    atoms = row.toatoms()
    
    positions = atoms.get_positions()
    x_min, y_min = positions[:, 0].min(), positions[:, 1].min()
    x_max, y_max = positions[:, 0].max(), positions[:, 1].max()
    z_max = positions[:, 2].max()
    
    center_x = 0.5 * (x_min + x_max)
    center_y = 0.5 * (y_min + y_max)
    
    atoms_ads = atoms.copy()
    add_adsorbate(atoms_ads, adsorbate, z_ads_1, (center_x, center_y))

    model_name = row.formula if row.formula else f"model_{row.id}"
    db_ads.write(atoms_ads, model=model_name)
    
    print(f"Added Li on top of model {model_name} (id={row.id}).")
```
